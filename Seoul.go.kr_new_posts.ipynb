```python
# make these to function
def get_content():
    news = ch_driver.find_elements_by_xpath(a_path)
    result = []
    for item in news:
        # get href
        news_link = item.get_attribute('href')
        # print(news_link)

        # get subject, post id
        subject_id = item.find_element_by_class_name('subject').text.split(']')
        post_id = int(subject_id[0].strip('['))
        subject = subject_id[1].strip()
        # print(post_id, subject)

        # get post date, time, category
        date_cat = item.find_element_by_class_name('date').text.split()
        post_date_time = (date_cat[0],date_cat[1])
        post_cat = date_cat[2].strip('[]')
        # print(post_date_time, post_cat)

        # get abstraction
        abs_text = item.find_element_by_class_name('txt').text
        # print(abs_text)

        result.append(
            (news_link, 
             post_id, 
             subject, 
             post_date_time, 
             post_cat, 
             abs_text,))
    
    return result
```

```python
# ch_driver.execute_script('fnPagingMove(2);') => can move to page 2
# so, fnPagingMove(n) => 1 to 10
from time import sleep


# tell driver to wait for 5 secs when certain elements are not available immediately.
ch_driver.implicitly_wait(5)

all_result = []
for i in range(1,10+1):
    # move to page i
    ch_driver.execute_script('fnPagingMove({page_num});'.format(page_num=i))
    # after 1 sec sleep,
    sleep(1)
    # get all news_content, append to all_result
    all_result.append(get_content())
    
all_result
```
