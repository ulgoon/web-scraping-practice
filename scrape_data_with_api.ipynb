```python
from selenium import webdriver
```

```python
ch_driver = webdriver.Chrome('/Users/ulgoon/Documents/dev/chromedriver')
```

```python
news_list_url = "http://seoul.go.kr/realmnews/in/list.do"
ch_driver.get(news_list_url)
```

```python
news_item_xpath = '//*[@id="content"]/div[2]/div/div[1]/div/a'
```

```python
results = ch_driver.find_elements_by_xpath(news_item_xpath)
len(results)
```

```python
for item in results:
    subject = item.find_element_by_class_name('subject').text
    date = item.find_element_by_class_name('date').text
    abs_text = item.find_element_by_class_name('txt').text
    print(subject, date, abs_text, sep='\n')
```

```python
def get_news_list():
    result_list = []
    results = ch_driver.find_elements_by_xpath(news_item_xpath)
    for item in results:
        subject = item.find_element_by_class_name('subject').text
        date = item.find_element_by_class_name('date').text
        abs_text = item.find_element_by_class_name('txt').text
        print(subject, date, abs_text, sep='\n')
        result_list.append((
            subject,
            date,
            abs_text,
        ))
    return result_list
```

```python
ch_driver.execute_script("fnPagingMove(1);")
```

```python
from time import sleep


ch_driver.implicitly_wait(5) # 요소가 로드되지 않았을 경우, 5초만큼 기다림

all_result = []
for i in range(1,10+1):
    ch_driver.execute_script("fnPagingMove({page_num});".format(
        page_num=i))
    item_data = get_news_list()
    all_result.append(tuple(item_data))
    sleep(1)
```
